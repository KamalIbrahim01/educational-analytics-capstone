## 02 - Data Extraction and Anonymization

### Objective

The main purpose of this notebook is to extract the all relevant student performance data from the three academic term workbooks in csv files and prepare them for analysis.

The notebook performs the following tasks:
1. Inspect the workbook structure.
2. Identify the worksheet(s) containing student performance data.
3. Extract the selected worksheet(s) into pandas DataFrames.
4. Replace personally identifiable information (PII) i.e student names with anonymous identifiers (Student ID).
5. Export the anonymized datasets as CSV files for subsequent cleaning and analysis.

No data cleaning or transformation is performed in this notebook beyond anonymization.

```python
import pandas as pd
from openpyxl import load_workbook
```

```python
# Detecting File paths
term1 = "../data/raw/SMS B7 TM 1.xlsm"
term2 = "../data/raw/SME BS7 T2.xlsm"
term3 = "../data/raw/SME BS7 (T3).xlsm"
```

```python
# Load the workbooks
wb_term1 = load_workbook(term1, data_only=True)
wb_term2 = load_workbook(term2, data_only=True)
wb_term3 = load_workbook(term3, data_only=True)
```

### Inspecting workbook Structure
The worksheet names are displayed to identify the sheets containing student performance records.

```python
workbooks = {
    "Term 1": wb_term1,
    "Term 2": wb_term2,
    "Term 3": wb_term3
}

for name, workbook in workbooks.items():

    print("=" * 50)
    print(name)
    print("=" * 50)

    for sheet in workbook.sheetnames:
        print(sheet)

    print()
```

## Extracting sheet to csv

Not every worksheet in the workbooks is required for analysis. Dashboard, template and reference sheets are excluded because they do not contribute to the analytical objectives of the project.

The following worksheets will be extracted and converted into CSV files for subsequent anonymization and data cleaning.
1. Students
2. Subjects
3. Attendance
4. Test 1
5. Test 2
6. Group Work
7. Project Work
8. Examscore
9. Class Score
10. General Remarks

```python
# Raw workbook paths

workbooks = {
    "term1": "../data/raw/SMS B7 TM 1.xlsm",
    "term2": "../data/raw/SME BS7 T2.xlsm",
    "term3": "../data/raw/SME BS7 (T3).xlsm"
}
```

```python
# Worksheets required for analysis

sheets = [
    "Student",
    "Subjects",
    "Attendance",
    "Test 1",
    "Test 2",
    "Group Work",
    "Project Work",
    "Examscore",
    "ClassScore",
    "General Remarks"
]
```

```python
# Extract worksheets and save as CSV

for term, workbook in workbooks.items():

    print(f"\nProcessing {term}...")

    for sheet in sheets:

        try:

            df = pd.read_excel(
                workbook,
                sheet_name=sheet
            )

            filename = sheet.lower().replace(" ", "_")

            output_path = f"../data/interim/{term}_{filename}.csv"

            df.to_csv(
                output_path,
                index=False
            )

            print(f"✓ Saved: {output_path}")

        except Exception as e:

            print(f"✗ Failed to extract '{sheet}' from {term}")
            print(e)
```

## Extraction Summary

The extraction process converts the selected worksheets from each academic term workbook into CSV files.

These CSV files will serve as the working datasets for the remainder of the project. The original Excel workbooks remain unchanged, preserving the integrity of the raw data.

The extracted files will be anonymized and cleaned in the next notebook before being combined into a single analytical dataset.
