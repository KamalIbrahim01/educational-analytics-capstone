## 01 - Data Inventory

### Objectives

The purpose of this notebook is to perform an initial deep dive inventory of the raw datasets used in the Educational Analytics Capstone project.

This notebook would document the structure and quality of the three academic term workbooks before any anonymization, cleaning, or transformation is performed.

This notebook will:

- Identify all worksheets contained in each workbook.
- Examine dataset dimensions.
- Review column names and data types.
- Detect missing values.
- Identify duplicate records.
- Identify personally identifiable information (PII).
- Produce an inventory report to guide subsequent stages of the project.

```python
import pandas as pd 
from openpyxl import load_workbook
```

```python
term1 = "../data/raw/SMS B7 TM 1.xlsm"
term2 = "../data/raw/SME BS7 T2.xlsm"
term3 = "../data/raw/SME BS7 (T3).xlsm"
```

```python
wb_term1 = load_workbook(term1, data_only=True)
wb_term2 = load_workbook(term2, data_only=True)
wb_term3 = load_workbook(term3, data_only=True)
```

### Listing the worksheets in the workbook

```python
workbooks = {
    "Term 1": wb_term1,
    "Term 2": wb_term2,
    "Term 3": wb_term3
}

for term, wb in workbooks.items():
    print(f"\n{'='*50}")
    print(term)
    print(f"{'='*50}")

    for sheet in wb.sheetnames:
        print(sheet)
```

### A summary of the workbooks 

```python
summary = []

for term, wb in workbooks.items():

    summary.append({
        "Workbook": term,
        "Number of Worksheets": len(wb.sheetnames)
    })

summary_df = pd.DataFrame(summary)

summary_df
```
